Exploratery Data Analysis EDA

```python
ax=sns.countplot(x="Gender",data=df)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_gn=df.groupby(["Gender"],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False)
sns.barplot(x="Gender",y="Amount",data=sales_gn)
```

Frome above two garph we have seen that there are large amount of shopping done by the women In Both purchase and the Orders 

```python
df.columns
```

```python
ax=sns.countplot(data=df,x="Age Group",hue="Gender")
```

```python
sales_age=df.groupby(["Age Group"],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False)
sns.barplot(x="Age Group",y="Amount",data=sales_age)    
```

From the above Both graph it's seen that there is the larger amount of the customer of the age Bracket 26-35

```python
df.columns
```

```python
sales_state=df.groupby(["State"],as_index=False)["Orders"].sum().sort_values(["Orders"],ascending=False).head(7)
sns.set(rc={"figure.figsize":(15,5)}, palette="pastel")
sns.barplot(data=sales_state,x="State",y="Orders")
```

```python
df.columns
```

```python
state_amount=df.groupby(["State"],as_index=False)["Amount"].sum().sort_values("Amount",ascending=False).head(8)
sns.set(rc={"figure.figsize":(15,5)})
sns.barplot(data=state_amount,x="State",y="Amount")
```

Frome above graphs we can see that there is the most of the order and Total Sales/amount are from the uttrapradesh , mahrashtra , karnataka , delhi 

```python
df.columns
```

Maritial Status

```python
ax=sns.countplot(data=df,x="Marital_Status")
sns.set(rc={"figure.figsize":(7,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
Sales_state=df.groupby(["Marital_Status","Gender"],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False)
sns.set(rc={"figure.figsize":(7,5)},palette="pastel",style="darkgrid")
sns.barplot(data=Sales_state,x="Marital_Status",y="Amount",hue="Gender")
```

From the above graph its states that the maximum of amount and purchase is done by the Marid Womens

```python
df.columns
```

```python
sns.set(rc={"figure.figsize":(20,5)},palette="muted")
ax=sns.countplot(data=df,x="Occupation",palette="pastel",hue="Occupation")
for basr in ax.containers:
    ax.bar_label(bars)
```

```python
sales_Occupation=df.groupby(["Occupation"],as_index=False)["Amount"].sum().sort_values("Amount",ascending=False)
sns.set(rc={"figure.figsize":(20,5)},palette="pastel")
sns.barplot(data=sales_Occupation,x="Occupation",y="Amount")
```

As from the Above Graph its seen that there is a large amount of Occupation and purchasing power is of IT,Healthcare,Aviation 

*Product_Category*

```python
df.columns
```

```python
ax=sns.countplot(data=df,x="Product_Category")
sns.set(rc={"figure.figsize":(25,5)})
for basr in ax.containers:
    ax.bar_label(bars)
```

```python
sales_Product_Category=df.groupby(["Product_Category"],as_index=False)["Amount"].sum().sort_values(by=["Amount"],ascending=False).head(7)
sns.set(rc={"figure.figsize":(25,5)})
sns.barplot(data=sales_Product_Category,x="Product_Category",y="Amount")
```

Frome above graph we have seen that most of the product are from the food, clothing and the electronics 

```python
df.columns
```

```python
Sales_product=df.groupby(["Product_ID"],as_index=False)["Orders"].sum().sort_values(by="Orders",ascending=False).head(7)
sns.set(rc={"figure.figsize":(20,5)})
sns.barplot(data=Sales_product,x="Product_ID",y="Orders")
```

Married women from age group 26-35 from UP, Mahrsahta ,Karnatka ,working in IT,Healthcare and aviation are more likely by products from Food ,clothes,Electronics Category
